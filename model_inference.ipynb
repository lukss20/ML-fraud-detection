```python
import mlflow
import dagshub

dagshub.init(
    repo_owner="luchit22",
    repo_name="ML-fraud-detection",
    mlflow=True
)
```

```python
model_uri = "models:/XGBoost_Fraud_Model/1"

model = mlflow.sklearn.load_model(model_uri)

print("Model loaded successfully")
```

```python
test_transaction = pd.read_csv("/kaggle/input/competitions/ieee-fraud-detection/test_transaction.csv")
test_identity = pd.read_csv("/kaggle/input/competitions/ieee-fraud-detection/test_identity.csv")
```

```python
test = test_transaction.merge(
    test_identity,
    on="TransactionID",
    how="left"
)
```

```python
preds = model.predict_proba(test)[:, 1]
```

```python
submission = pd.DataFrame({
    "TransactionID": test["TransactionID"],
    "isFraud": preds
})
```

```python
submission.to_csv("submission.csv", index=False)

submission.head()
```
